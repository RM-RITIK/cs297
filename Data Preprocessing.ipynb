```python
import pandas as pd
data = pd.DataFrame(columns = ['OpCode Sequence', 'Malware Family'])
```

```python
directory = 'Dataset'
for folder_name in os.listdir(directory):
    folder = os.path.join(directory, folder_name)
    for file_name in os.listdir(folder):
        file = os.path.join(folder, file_name)
        ext = os.path.splitext(file)[-1].lower()
        if ext == '.txt':
            with open(file) as f:
                seq = ""
                for line in f.readlines():
                    line = line.strip()
                    seq = seq + line + " "
                seq = seq.strip()
                data = data.append({'OpCode Sequence': seq, 'Malware Family': folder_name}, ignore_index = True)
```

```python
family = []
num_samples = []
for cls in pd.unique(data['Malware Family']):
    family.append(cls)
    num_samples.append(len(data.loc[data['Malware Family'] == cls]))
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize = (30, 10))
plt.bar(family, num_samples)
plt.xticks(range(len(family)), family, rotation='vertical')
plt.show()
```
